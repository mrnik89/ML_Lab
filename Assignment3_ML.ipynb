```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

```

```python
df = pd.read_csv("loan_prediction.csv")
```

```python
print("Dataset Shape:", df.shape)
df.info()
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
df.describe(include='all')
```

```python
for col in ['Gender', 'Married', 'Dependents', 'Self_Employed', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].median())
```

```python
print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
df.drop('Loan_ID', axis=1, inplace=True)

label_cols = ['Gender', 'Married', 'Dependents', 'Education',
              'Self_Employed', 'Property_Area', 'Loan_Status']

le = LabelEncoder()
for col in label_cols:
    df[col] = le.fit_transform(df[col])
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

scaled_df = pd.DataFrame(scaled_data, columns=df.columns)
scaled_df.head()
```

```python
inertia = []
K = range(1, 15)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_df)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
scaled_df['Cluster'] = kmeans.fit_predict(scaled_df)

# Visualize using PCA
pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_df.drop("Cluster", axis=1))

plt.figure(figsize=(8,6))
sns.scatterplot(x=pca_components[:,0], y=pca_components[:,1], hue=scaled_df['Cluster'], palette='Set1')
plt.title('Loan Applications Clusters via KMeans + PCA')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```
