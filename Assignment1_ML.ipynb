```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/train.csv')
```

```python
print(df.head())
```

```python
df.info()
```

```python
print(df.describe())
```

```python

# Check for missing values

print(df.isnull().sum())
```

```python
# Check for duplicate rows
print(f"Duplicate rows: {df.duplicated().sum()}")
```

```python
# Drop duplicate rows if any
df.drop_duplicates(inplace=True)
```

```python
# List of categorical columns
categorical_cols = ['Gender', 'Vehicle_Age', 'Vehicle_Damage', 'Driving_License', 'Previously_Insured', 'Policy_Sales_Channel', 'Region_Code']
```

```python
# Plot count plots for categorical variables
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=col, data=df, palette='pastel')
    plt.title(f'Distribution of {col}', fontsize=14)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# List of numerical columns
numerical_cols = ['Age', 'Annual_Premium', 'Vintage']
```

```python
# Plot histograms for numerical variables
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, color='skyblue', bins=30)
    plt.title(f'Distribution of {col}', fontsize=14)
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

```python
# Scatter plot: Age vs. Annual Premium
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='Annual_Premium', data=df, hue='Response', palette='coolwarm', alpha=0.6)
plt.title('Age vs. Annual Premium by Response', fontsize=14)
plt.xlabel('Age')
plt.ylabel('Annual Premium')
plt.tight_layout()
plt.show()
```

```python
# Boxplot: Vehicle Age vs. Annual Premium
plt.figure(figsize=(8, 5))
sns.boxplot(x='Vehicle_Age', y='Annual_Premium', data=df, palette='Set2')
plt.title('Vehicle Age vs. Annual Premium', fontsize=14)
plt.xlabel('Vehicle Age')
plt.ylabel('Annual Premium')
plt.tight_layout()
plt.show()
```

```python
# Compute correlation matrix
corr_matrix = df[numerical_cols + ['Response']].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Pairplot for numerical variables
sns.pairplot(df[numerical_cols + ['Response']], hue='Response', palette='husl', diag_kind='kde')
plt.show()
```
